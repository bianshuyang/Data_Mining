```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'df' with a 'Date' column and 'Description' column

start_year = 2008
end_year = 2023

for year in range(start_year, end_year + 1):
    try:
        df_year = df_small.loc[f'{year}-01-01':f'{year}-12-31']['Description']
        df_year = df_year[df_year.notnull()]
        df_year_words = df_year.values
        df_year_words_no_space = ' '.join(list(df_year_words))
        st = set(STOPWORDS)
        uninteresting = ['year','rejected','letter','rejection','acceptance','email','e-mail','one','school','check','application','website','interview','funding','email','got','accepted','decision','admission','offer','will','received','program']
        for w_unint in uninteresting:
            st.add(w_unint)
        wordcloud = WordCloud(width=800, height=800,
                              background_color='white',
                              stopwords=st,
                              min_font_size=10).generate(df_year_words_no_space)

        plt.figure(figsize=(3, 3), facecolor=None)
        plt.imshow(wordcloud)
        plt.axis("off")
        plt.tight_layout(pad=0)
        plt.title(f"Word Cloud for {year}")
        plt.show()
    except:
        print(year)
```

```python
import pandas as pd
from fuzzywuzzy import process

# Assuming df_small is your DataFrame and 'Major' is the column with majors

# Step 1: Load the list of STEM OPT eligible majors
with open('STEMopt.txt', 'r') as f:
    stem_opt_eligible = [line.strip() for line in f.readlines()]

# Step 2: Use value_counts to find the frequency of each major
majors_frequency = df_small['Major'].value_counts()
majors_frequency  = majors_frequency[majors_frequency > 100]

# Initialize lists for verified majors
verified_stem_opt_eligible = []

# Threshold for automatic classification (e.g., 90 out of 100)
threshold_score = 90

# Limit for human verification
verification_limit = 5
verification_count = 0

# Step 3: Classify majors with fuzzy matching
for major in tqdm(majors_frequency.index, desc="Classifying Majors"):
    match, score = process.extractOne(major, stem_opt_eligible)
    if score >= threshold_score:
        verified_stem_opt_eligible.append(major)
    elif verification_count < verification_limit:
        # For low-confidence matches, prompt for human verification
        print(f"Major: {major}, Best Match: {match}, Score: {score}")
        user_input = input("Is this major STEM OPT eligible? (y/n): ")
        if user_input.lower() == 'y':
            verified_stem_opt_eligible.append(major)
            
        if verification_count >= verification_limit:
            print("Verification limit reached. Proceeding with verified majors.")
            break
        verification_count += 1
# Filter the original DataFrame to get rows with verified STEM OPT eligible majors


# df_stem_opt_eligible now contains rows of df_small that are STEM OPT eligible
```

```python
df_stem_opt_eligible = df_small[df_small['Major'].isin(verified_stem_opt_eligible)]

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming df_stem_opt_eligible is your initial DataFrame and it's already loaded
df_updated = df_stem_opt_eligible.copy()
df_updated = df_updated.dropna(subset=['GPA'])
df_updated['GPA'] = pd.to_numeric(df_updated['GPA'], errors='coerce')
df_updated = df_updated[df_updated['GPA'] <= 4]

# Step 1: Categorize GPAs into clusters
def categorize_gpa(gpa):
    if 2 <= gpa < 3:
        return '2-3'
    elif 3 <= gpa < 3.3:
        return '3.0-3.3'
    elif 3.3 <= gpa < 3.5:
        return '3.3-3.5'
    elif 3.5 <= gpa < 3.7:
        return '3.5-3.7'
    elif 3.7 <= gpa <= 4:
        return '3.7-4.0'
    else:
        return 'Unknown'  # For GPAs outside the specified ranges

df_updated['GPA_Cluster'] = df_updated['GPA'].apply(categorize_gpa)

# Step 2 & 3: Group by GPA clusters, citizenship, and status, then count
gpa_status_counts = df_updated.groupby(['GPA_Cluster', 'Citizenship', 'Status']).size().unstack(fill_value=0)

# Step 4: Visualize the results
# You can plot the counts directly, but here's how to create a stacked bar chart for better visualization
gpa_status_counts.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('GPA Clusters by Citizenship and Status')
plt.ylabel('Count')
plt.xlabel('GPA Cluster and Citizenship')
plt.xticks(rotation=45)
plt.legend(title='Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Calculate the total applications for each GPA and citizenship category
gpa_status_counts['Total'] = gpa_status_counts.sum(axis=1)

# Calculate the acceptance rate
gpa_status_counts['Acceptance Rate'] = gpa_status_counts['Accepted'] / gpa_status_counts['Total']

# Reset index if needed to use 'GPA_Cluster' and 'Citizenship' as plotting parameters
gpa_status_counts_reset = gpa_status_counts.reset_index()

# Create a grouped bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='GPA_Cluster', y='Acceptance Rate', hue='Citizenship', data=gpa_status_counts_reset)

plt.title('Acceptance Rate by GPA Cluster and Citizenship')
plt.xlabel('GPA Cluster')
plt.ylabel('Acceptance Rate')
plt.xticks(rotation=45)
plt.legend(title='Citizenship')
plt.tight_layout()
plt.show()
```
